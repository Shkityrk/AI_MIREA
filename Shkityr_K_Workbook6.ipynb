–ü–æ—Å–ª–µ–¥–Ω—è—è —Ö—Ä–æ–º–æc–æ–º–∞ (—Å –Ω–∏–∑—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º) –≤—ã–±—ã–≤–∞–µ—Ç –∏–∑ –ø–æ–ø—É–ª—è—Ü–∏–∏. –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –≤ –ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏.
–ü–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ù–∞–ø–∏—à–µ–º –∏—Ö.

```python
def qZ(x,y):
    return (x-3 * y+1)/(3*x**2+3*y**2+1)

def qSumZ(Z):
    return sum(Z)

def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]
    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]
    X[0] = oldX[sortedId[0]]
    X[1] = oldX[sortedId[1]]
    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]
    Y[2] = oldY[sortedId[0]]
    Y[3] = oldY[sortedId[1]]
    return X, Y

def sorting(Z):
    sortedId=sorted(range(len(Z)), key=lambda k:Z[k])
    return sortedId

def evoStep(X,Y,Z):
    _,minId = min((value, id) for (id, value) in enumerate(Z))
    X=X[:]
    Y=Y[:]
    Z=Z[:]
    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)
    
    return X,Y,Z
```

```python
def evosteps(X, Y, stepsnum=4):
    results=[]

    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]
        X, Y, Z = evoStep(X, Y, arrZ)
        X, Y = exchangeScheme(X, Y, sorting(Z))

        results.append([X, Y, qSumZ(arrZ), arrZ])

    return X,Y,results
```

```python
X = [-2,-1,0,1]
Y = [-2,-1,0,1]

results = evosteps(X,Y)
```

```python
for i in range(len(results[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')
qualityArrZ =[]
for i in range(len(results[2])): 
    qualityArrZ += results[2][i][3]
print(f'max Z:{max(qualityArrZ)}')
```

# –ó–∞–¥–∞–Ω–∏–µ:
–í—ã–ø–æ–ª–Ω–∏—Ç–µ –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –≥–µ–Ω–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–ª–æ–∂–µ–Ω–Ω—ã–º–∏ –Ω–∞—á–∞–ª—å–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏.

```python
X = [-1, 0, 2, 3]
Y = [0, 1, -2, 2]

def qetZ(x,y):
    return (x-3*y)/(2*x**2 + 2*y**2 + 1)

def evoSteps_task(X, Y, stepsNum = 4): 
    results = []
    for i in range(4):
        arrZ = [qetZ(x, Y[i]) for i, x in enumerate(X)]
        X, Y, Z = evoStep(X, Y, arrZ)
        X, Y = exchangeScheme(X, Y, sorting(Z))
        results.append([X, Y, qSumZ(arrZ), arrZ])
    return X, Y, results
```

```python
res = evoSteps_task(X,Y)
```

```python
for i in range(len(res[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')
qualityArrZ =[]
for i in range(len(res[2])): 
    qualityArrZ += res[2][i][3]
print(f'max Z:      {max(qualityArrZ)}')
```

# 1.2.1 –ü—Ä–∏–º–µ—Ä

```python
import networkx as nx
from math import e

distances =[(1,2,20),
            (1,3,40),
            (1,4,42),
            (1,5,33),
            (1,6,21),
            (2,3,26),
            (2,4,38),
            (2,5,42),
            (2,6,17),
            (3,4,22),
            (3,5,43), 
            (3,6,21), 
            (4,5,27),
            (4,6,22),
            (5,6,26)
            ]

V = [1, 4, 5, 2, 6, 3, 1] # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3, 4),
     (4, 6),
     (5, 2), 
     (6, 2)] # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P= [49, 54, 43, 54] # —Å–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á–µ—Ç–∞
T= 100 # –Ω–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
```

```python
def probability(delta, T):
    return 100*e**(-delta / T)

# —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã 
def reductTemp(prevT):
    nextT = 0.5 * prevT
    return nextT

graph = nx.Graph()
graph.add_weighted_edges_from(distances)

nx.draw_kamada_kawai(graph, 
                     node_color= '#fb7258', 
                     node_size=2000,
                     with_labels=True)
```

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip: return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])

def routeLength(V, distances): 
    edges = []
    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))
    return sum(edges)

def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0
    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]
    return arrV

def routeReplacement(V,Z):
    for z in Z:
        V=routeOneReplacement(V,z)
    return V

# –≤—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã numa
    arrSum = [sumLength] # –º–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω
    # —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances) # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ 
        arrSum.append(newS)
        deltas = newS - sumLength # —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–æ–≤
        # –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 
        if deltas > 0:
            p = probability(deltas, T) # –ø–æ–¥—Å—á—ë—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    # –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ —Ä, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è
            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T)
    return V, arrSum

def drawRouteGraph(distances, bestRoute):
    newDistances =[]
    # –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1]or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                    newDistances.append(distance)
   
    graph = nx.Graph() # —Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
    graph.add_weighted_edges_from(newDistances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä
    # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)
    
    
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)
print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}') 
print(f'–î–ª–∏–Ω—ã –≤—Å–µx p–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')
drawRouteGraph(distances, bestRoute) # –æmpu—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```

# –ó–∞–¥–∞–Ω–∏–µ
–ù–∞–π—Ç–∏ –¥–ª–∏–Ω—É –≥–∞–º–∏–ª—å—Ç–æ–Ω–æ–≤–∞ —Ü–∏–∫–ª–∞ S4 –≤ –ø–æ–ª–Ω–æ–º –≥—Ä–∞—Ñ–µ K6 –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —Ü–∏–∫–ª–æ–≤ —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞ –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞–º –Ω–∏–∂–µ.

```python
distances=[(1,2,19),
           (1,3,39), 
           (1,4,35), 
           (1,5,26),
           (1,6,18),
           (2,3,26),
           (2,3,26),
           (2,4,33),
           (2,5,37),
           (2,6,14), 
           (3,4,22), 
           (3,5,41), 
           (3,6,21),
           (4,5,22),
           (4,6,19), 
           (5,6,24)]

V=[1,4,2,3,5,6,1]
Z=[(5,2),
   (4,5),
   (2,3), 
   (3,4)]

P=[88,54,24,64]
```

```python
graph = nx.Graph()
graph.add_weighted_edges_from(distances)

nx.draw_kamada_kawai(graph, 
                     node_color= '#fb7258', 
                     node_size=2000,
                     with_labels=True)
```

```python
# –≤—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã numa
    arrSum = [sumLength] # –º–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω
    # —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances) # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ 
        arrSum.append(newS)
        deltas = newS - sumLength # —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–æ–≤
        # –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–∞–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 
        if deltas > 0:
            p = probability(deltas, T) # –ø–æ–¥—Å—á—ë—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    # –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ —Ä, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è
            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T)
    return V, arrSum
```

```python
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)
print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}') 
print(f'–î–ª–∏–Ω—ã –≤—Å–µx p–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')
drawRouteGraph(distances, bestRoute) # –æmpu—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```
